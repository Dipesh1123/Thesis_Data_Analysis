```python
#MAPPING THE VALUE

mapping = {
    'Not Included': 0,
    'Partially included (mentioned in course)': 1,
    'Fully included (dedicated  course/module)': 2
}

df['Prior_exposure_curriculum'] = df['Inclusion of digital health in your curriculum:'].map(mapping)

mapping_training = {
    'No formal training': 0,
    'Yes (Participated but not certified)': 1,
    'Yes (certified training)': 2
}

df['Prior_Exposure_Training'] = df['Formal Training in digital health'].map(mapping_training)

mapping_training = {
    'No Any Courses': 0,
    'Completed 1 course': 1,
    'Completed ≥2 courses': 2
}

df['Prior_Exposure_online_course'] = df['Online Courses on Digital Health:'].map(mapping_training)
```

```python
# Sum the mapped prior exposure columns
df['Total_Prior_Exposure'] = df['Prior_exposure_curriculum'] + df['Prior_Exposure_Training'] + df['Prior_Exposure_online_course']

# Define the exposure categories based on the sum
def categorize_exposure(total_exposure):
    if total_exposure <= 2:
        return 'Low Exposure'
    elif 3 <= total_exposure <= 4:
        return 'Moderate Exposure'
    elif total_exposure >= 5:
        return 'High Exposure'
    else:
        return 'Unknown' # Handle any unexpected values

# Apply the categorization function to create the new column
df['Prior_Exposure_Category'] = df['Total_Prior_Exposure'].apply(categorize_exposure)



exposure_summary = pd.DataFrame({
    'Count':df['Prior_Exposure_Category'].value_counts() ,
    'Percentage': (df['Prior_Exposure_Category'].value_counts() / len(df)) * 100})

display(exposure_summary)
```

```python
from scipy.stats import ttest_1samp

# Perform one-sample t-test on 'Total_knowledge' against a benchmark of 50%
# Since 'Total_knowledge' is a sum out of 25 questions, the benchmark is 25 * 0.50 = 12.5
benchmark = 12.5
t_statistic, p_value = ttest_1samp(df['Total_knowledge'], benchmark)

display(f"One-Sample T-test against a benchmark of {benchmark} (50%):")
display(f"T-statistic: {t_statistic}")
display(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    display(f"Result: The mean total digital health knowledge score is significantly different from the benchmark of {benchmark} (p < {alpha}).")
else:
    display(f"Result: The mean total digital health knowledge score is not significantly different from the benchmark of {benchmark} (p >= {alpha}).")
```

```python
from scipy.stats import chi2_contingency
import numpy as np

# 1. Create a contingency table for 'Name of University (eg. Purbanchal University)' and 'Knowledge_Classification'
contingency_table_observed_uni = pd.crosstab(df['Name of University (eg. Purbanchal University)'], df['Knowledge_Classification'])
display("Observed Contingency Table (University vs. Knowledge Classification):")
display(contingency_table_observed_uni)

# 2. Calculate the observed Chi-square statistic
chi2_observed_uni, p_observed_uni, dof_observed_uni, expected_observed_uni = chi2_contingency(contingency_table_observed_uni)

display(f"\nObserved Chi-square statistic: {chi2_observed_uni:.2f}")

# 3. Initialize list for permuted Chi-square statistics
permuted_chi2_statistics_uni = []

# 4. Set the number of permutations
n_permutations = 5000

# 5. Loop for permutations
for _ in range(n_permutations):
    # a. Randomly shuffle 'Knowledge_Classification'
    df_shuffled_uni = df.copy()
    df_shuffled_uni['Knowledge_Classification_permuted'] = np.random.permutation(df_shuffled_uni['Knowledge_Classification'])

    # b. Create a new contingency table with permuted data
    contingency_table_permuted_uni = pd.crosstab(
        df_shuffled_uni['Name of University (eg. Purbanchal University)'],
        df_shuffled_uni['Knowledge_Classification_permuted']
    )

    # c. Calculate the Chi-square statistic for the permuted table
    chi2_permuted_uni, _, _, _ = chi2_contingency(contingency_table_permuted_uni)

    # d. Append the calculated Chi-square statistic to the list
    permuted_chi2_statistics_uni.append(chi2_permuted_uni)

# 6. Calculate the p-value from the permutation test
p_value_permutation_uni = np.sum(np.array(permuted_chi2_statistics_uni) >= chi2_observed_uni) / n_permutations

display(f"Permutation Test P-value (approximating Fisher-Freeman-Halton): {p_value_permutation_uni:.4f}")

alpha = 0.05
if p_value_permutation_uni < alpha:
    display(f"Result: There is a statistically significant association between University Name and Knowledge Classification (p < {alpha}).")
else:
    display(f"Result: There is no statistically significant association between University Name and Knowledge Classification (p >= {alpha}).")
```
